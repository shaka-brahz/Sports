```python
numSims = 50000

data = pd.read_excel('Test.xlsx', sheet_name='TeamStats')

data = data[['TEAM', 'POSS', 'STD_POSS', 'ORtg/POSS', 'DRtg/POSS', 'STD_ORtg', 'STD_DRtg']]

home_team = input('Enter the name of the home team: ')
away_team = input('Enter the name of the away team: ')

home_data = data.loc[data['TEAM'] == home_team]
away_data = data.loc[data['TEAM'] == away_team]
league_data = data.loc[data['TEAM'] == 'LEAGUE AVERAGE']

normPoss = (home_data.iloc[0]['POSS'] + away_data.iloc[0]['POSS'] + league_data.iloc[0]['POSS']) / 3
normStdPoss = (home_data.iloc[0]['STD_POSS'] + away_data.iloc[0]['STD_POSS'] + league_data.iloc[0]['STD_POSS']) / 3

possessions = []
for i in range(numSims):
    r = random.uniform(0,1)
    x = norm.pdf(r, loc=normPoss, scale=normStdPoss)
    possesions.append(x)
    
Poss/G = sum(possessions/len(possessions))
Poss/G
```
